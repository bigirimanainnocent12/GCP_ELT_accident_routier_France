```python
from google.cloud.storage import Bucket
import logging

logger = logging.getLogger(__name__)

def fichier_exists(bucket: Bucket, gcp_path: str) -> bool:
    """
    Vérifie si un fichier existe dans un bucket Google Cloud Storage.

    

    Raises:
        TypeError: Si les paramètres fournis ne sont pas du bon type.
        ValueError: Si le chemin GCS est vide.
    """

    # Validation des entrées
    if not isinstance(bucket, Bucket):
        raise TypeError("Le paramètre 'bucket' doit être une instance google.cloud.storage.Bucket.")

    if not isinstance(gcp_path, str):
        raise TypeError("Le paramètre 'gcp_path' doit être une chaîne de caractères.")

    if not gcp_path.strip():
        raise ValueError("Le chemin 'gcp_path' ne peut pas être vide.")

    # Log de debug
    logger.debug(f"Vérification de l'existence du fichier : gs://{bucket.name}/{gcp_path}")

    # Vérification de l'existence du fichier
    blob = bucket.blob(gcp_path)
    exists = blob.exists()

    logger.info(f"Fichier {'trouvé' if exists else 'non trouvé'} : gs://{bucket.name}/{gcp_path}")

    return exists
```
